```python
#if-else, when two outcomes

sales = float(input("Enter amount: "))
if sales > 10000:
    print("Target achieved")
else:
    print("Try more")
```

```python
#elif add when more then two condition
name = input("Your name: ")

maths = float(input("Enter marks of Maths: "))
Scince = float(input("Enter marks of Scince: "))
physics = float(input("Enter marks of Physics: "))
sst = float(input("Enter marks of SST: "))

total_student_marks = (maths + Scince + physics + sst)
total_marks = 400

percentage = round((total_student_marks / total_marks * 100), 2)

print("Percentage of total marks: ", percentage, "%")

if percentage >= 80:
     print("Grade A")
elif percentage >= 60: 
     print("Grade B")
elif percentage >= 30:
     print("Grade C")
else:
     print("tumse 30 number nhi laya gaya ... ",name)
```

```python
#Task- Discount calculation
bill = float(input("Enter bill amount: "))

if bill >= 5000:
    discount = bill * .20
    print("Discount applied 20%")
elif bill >= 3000:
    discount = bill * 0.10
    print("Discount applied 10%")
else:
    discount = 0
    print("Bad luck")

final_amount = bill - discount

print("Pay after discount: ", final_amount)
```
