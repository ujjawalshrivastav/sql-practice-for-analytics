```python
#Arithmetic operators 
Price = int(input("Enter product price: "))
Quantity = int(input("Enter quantity"))

total_amount = Price * Quantity
print("Total Amount: ", total_amount)
```

```python
#profit calculation 
selling_amount = float(input("Enter selling price: "))
cost_price = float(input("Enter cost price: "))

profit = selling_amount - cost_price

print("Profit: ", profit)
```

```python
#comparison
target = 10000
print("Target achieved: ", total_amount >= target)
```

```python
#Task 
salary = float(input("Enter monthly salary: "))
expenses = float(input("Enter monthly expenses: "))

savings = salary - expenses
print("Monthly Savings: ", savings)
print("Good savings: ", savings>5000)
```
